```python
df["總級分"] = df[["國文","數學","英文","自然","社會"]].sum(1)#一行一行相加
## 1是行 2是列
##新增一個總和的欄位：df["新增欄位名字"] = df[要總和的欄位的list].sum(1)
```

```python
df.head()
```

```python
 df["主科"] = df.數學*1.5+df.英文
##新增一個經過計算的欄位：df["新增欄位名字"] = df名稱.項目 計算式
```

```python
df.sort_values(by="總級分", ascending=False).head()
##依項目進行排序：
#由大到小：df名稱.sort_values(by="要排序項目", ascending=False)
#由小到大：df名稱.sort_values(by="要排序項目")
```
